# script_t1_02.py

ML problem: **classification**

This script loads the Iris dataset and splits it into train/test sets, then trains an XGBClassifier and makes predictions on the test set. It is a classification workflow using gradient boosting.

## Environment Configuration

```
from xgboost import XGBClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

## Data Preparation

```
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data['data'], data['target'], test_size=.2)
```

## Model Generation

```
# create model instance
bst = XGBClassifier(n_estimators=2, max_depth=2, learning_rate=1, objective='binary:logistic')
# fit model
bst.fit(X_train, y_train)
```

## Model Evaluation

```
# make predictions
preds = bst.predict(X_test)
```
